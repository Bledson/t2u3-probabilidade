```python
import math
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import linspace

blue = (31/255, 119/255, 180/255)
orange = (255/255, 127/255, 14/255)
green = (44/255, 160/255, 44/255)
red = (214/255, 39/255, 40/255)
purple = (148/255, 103/255, 189/255)
colors = [blue, orange, green, red, purple]
```

```python
def nmr(n, time, rate):
    return n * math.exp((n - 1) * rate * -time) - (n - 1) * math.exp(
        n * rate * -time)
```

```python
def mttf_nmr(n, rate):
    return (n + n - 1) / (n * (n - 1) * rate)
```

```python
def exp(time, rate, n=1):
    return 1 - math.pow((1 - math.exp(rate * -time)), n)
```

```python
def mttf_exp(rate, n=1):
    return 1 / (n * rate)
```

```python
def dists(dist, rate):
    tmr_values = [nmr(3, t, rate) for t in dist]
    qmr_values = [nmr(5, t, rate) for t in dist]
    p2_values = [exp(t, rate, 2) for t in dist]
    p3_values = [exp(t, rate, 3) for t in dist]
    p4_values = [exp(t, rate, 4) for t in dist]
    return [tmr_values, qmr_values, p2_values, p3_values, p4_values]
```

```python
r = 1
d = linspace(0, 2, 21)
dist_values = dists(d, r)

fig, axes = plt.subplots(5, 1, figsize=(5, 13))

fig.suptitle('Exponential distributions - λ = ' + str(r),
             fontsize=16)
fig.subplots_adjust(hspace=0.6)

for i, values in enumerate(dist_values):
    axes[i].plot(d, values, color=colors[i])
    axes[i].set_xlabel('R(t)')
    axes[i].set_ylabel('Probability')
    sns.despine(ax=axes[i])
    if i == 0:
        axes[i].set_title('TMR')
        axes[i].axvline(x=mttf_nmr(3, r), color='b')
    elif i == 1:
        axes[i].set_title('5MR')
        axes[i].axvline(x=mttf_nmr(5, r), color='b')
    elif i == 2:
        axes[i].set_title('Two-component Parallel')
        axes[i].axvline(x=mttf_exp(r, 2), color='b')
    elif i == 3:
        axes[i].set_title('Three-component Parallel')
        axes[i].axvline(x=mttf_exp(r, 3), color='b')
    elif i == 4:
        axes[i].set_title('Four-component Parallel')
        axes[i].axvline(x=mttf_exp(r, 4), color='b', label='MTTF')

plt.legend()
plt.show()
```
